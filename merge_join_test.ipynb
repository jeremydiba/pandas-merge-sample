```python
import pandas as pd
import time
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding='Latin-1').set_index("ORDERNUMBER")
```

```python
df.columns[:12]
```

```python
df_orders = df.iloc[:, : 13]
df_orders
```

```python
df_unique_customers = df.iloc[:, 12:]
df_unique_customers = df_unique_customers.groupby("CUSTOMERNAME").first()
df_unique_customers
```

```python
df_unique_customers_unindexed = df_unique_customers.reset_index()
```

Now that we have customers and orders, lets test join

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.join(df_unique_customers, on="CUSTOMERNAME", how="left")
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```

```python
df_combined
```

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.merge(df_unique_customers, how='left', on="CUSTOMERNAME")
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.merge(df_unique_customers_unindexed, how='left', on="CUSTOMERNAME")
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.merge(df_unique_customers, how='left', left_on="CUSTOMERNAME", right_index=True)
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.merge(df_unique_customers_unindexed, how='left', left_on="CUSTOMERNAME", right_on="CUSTOMERNAME")
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```
