```python
import pandas as pd
import time
import numpy as np
pd.set_option('display.max_columns', 6)
```

```python
df = pd.DataFrame(index=range(2823),columns=["ORDERNUMBER", "QUANTITYORDERED", "ITEMNAME", "PRICEEACH", "CUSTOMERNAME", "CUSTOMERADDRESS"])
df
```

```python
customers = {"Amazon":"303 West", "Walmart":"East ST", "Best Buy": "North ave", "State Farm": "North Highway", "Apple": "101 Circle", "Google": "300 Northeast", "McDonalds": "South Fayette", "Aldis": "East Bay", "Chick-fil-a": "Buffington RD", "Wendys": "Iowa RD", "Toyota": "100 East Main", "Volkswagen": "North Main", "Tesla": "Austin St", "Disney": "Orlando Rd", "Spiny.ai": "600 Nashville Rd", "Home Depot":"Fulton Rd"}
item_ordered = ["Phone", "Wallet", "Water", "Soda", "Computer", "Hard Drive", "TV", "Headphones", "Chicken Breast", "Tomatos", "Onions", "Ketchup", "Lettuce", "Blender", "Grill", "Mirowave"]
```

```python
df["ORDERNUMBER"] = df.index
df["QUANTITYORDERED"] = np.random.randint(1, 6, df.shape[0])
df["ITEMNAME"] = np.random.choice(list(item_ordered), len(df))
df["PRICEEACH"] = np.random.randint(1, 10, df.shape[0])
df["CUSTOMERNAME"] = np.random.choice(list(customers), len(df))
df["CUSTOMERADDRESS"] = df['CUSTOMERNAME'].map(customers)
df.set_index("ORDERNUMBER")
```

```python
df
```

```python
df.columns[:3]
```

```python
df_orders = df.iloc[:, : 4]
df_orders
```

```python
df_unique_customers = df.iloc[:, 4:]
df_unique_customers = df_unique_customers.groupby("CUSTOMERNAME").first()
df_unique_customers
```

```python
df_unique_customers_unindexed = df_unique_customers.reset_index()
```

Now that we have customers and orders, lets test join

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.join(df_unique_customers, on="CUSTOMERNAME", how="left")
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.merge(df_unique_customers, how='left', on="CUSTOMERNAME")
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.merge(df_unique_customers_unindexed, how='left', on="CUSTOMERNAME")
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.merge(df_unique_customers, how='left', left_on="CUSTOMERNAME", right_index=True, left_index=False)
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```

```python
start = time.time()
for i in range(10000):
    df_combined = df_orders.merge(df_unique_customers_unindexed, how='left', left_on="CUSTOMERNAME", right_on="CUSTOMERNAME")
end = time.time()
total_time = end - start
print("Time for 1000 joins")
print(end - start)
print("average time per join")
print(total_time / 10000)
```
